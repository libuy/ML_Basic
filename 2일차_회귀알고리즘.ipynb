```python
#넘파이
import numpy as np
import pandas as pd

#시각화
import matplotlib.pyplot as plt
import seaborn as sns #시각화 라이브러리! -> 덤 : plotly(https://plotly.com/examples/)
```

```python
#값은 동일하지만, 시각화를 위해 데이터의 형태를 바꾼 것
X = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0,
     6.0, 4.0, 12.0, 7.0, 5.0]

# 종속 변수
y = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96,
     7.24, 4.26, 10.84, 4.82, 5.68]
```

```python
#데이터 프레임 제작(zip을 통해 X와 y를 하나로 묶어줌)
df = pd.DataFrame(zip(X,y), columns=['x', 'y'])
df.head()
```

```python
#산점도(점으로 나타내는 그래프) sns.scatterplot
#데이터 프레임의 형태로 그리려는 데이터를 입력
sns.scatterplot(data=df, x=df['x'], y=df['y'])
```

# 1. 선형 회귀

```python
# 가상 데이터
X = [[10.0], [8.0], [13.0], [9.0], [11.0], [14.0],
     [6.0], [4.0], [12.0], [7.0], [5.0]]
```

```python
from sklearn.linear_model import LinearRegression

#y = ax+b -> a가 뭐야? b가 뭐야?
linear = LinearRegression()
linear.fit(X, y)
```

```python
#기울기
print(linear.coef_)
#절편
print(linear.intercept_)

# y = 0.50009091x + 3.0000909090909094
```

```python
#예측
#y = 0.50009091x + 3.0000909090909094 < 와 같이 선형 회귀 모델의 계수와 절편을 찾았으니 -> 14를 넣으면 얼마나 나올지 계산해라.
print( linear.predict([[14]])  )
```

```python
0.50009091 * 14 + 3.0000909090909094
```

```python
sns.scatterplot(df, x=df['x'], y=df['y'])

#내가 찾은 선형 회귀 모델을 그래프에 나타내보자!
#직선 그래프 -> lineplot
sns.lineplot(x=[0, 14], y=[3.0000909090909094, 10.0013636])
```

# 라쏘, 릿지, 엘라스틱 넷

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet

#정의
las = Lasso()
rid = Ridge()
ela = ElasticNet()

#훈련
las.fit(X, y)
rid.fit(X, y)
ela.fit(X, y)

#예측
```

```python
print(las.coef_, las.intercept_)
print(rid.coef_, rid.intercept_)
print(ela.coef_, ela.intercept_)
```

```python
sns.scatterplot(df, x=df['x'], y=df['y'])

data = [[0], [14]]

#lasso모델에게, 0과 14를 주고 예측해보라고 함
#결과값은 y_pred
y_pred = las.predict(data)
sns.lineplot(x=[ 0, 14], y=y_pred, label='lasso', linestyle='-')

y_pred = rid.predict(data)
sns.lineplot(x=[ 0, 14], y=y_pred, label='ridge', linestyle='--')

y_pred = ela.predict(data)
sns.lineplot(x=[ 0, 14], y=y_pred, label='ela', linestyle=':')

plt.legend()
plt.tight_layout()
plt.show()
```
