#데이터 프레임을 특정 조건에 따라 '집계'

```python
#연도별 기대수명의 평균
train.head()
```

```python
#groupby(기준)[무엇을].어떻게(함수)
train.groupby('year')['lifeExp'].mean()
```

```python
#대륙을 기준으로 인구 평균
train.groupby('continent')['population'].mean()
```

```python
#40분까지!!
#1. 대륙을 기준으로 (중복없는..)나라의 갯수를 세시오.
#unique() : 중복없는 값 자체
#nunique() : 중복없는 값의 '갯수'
train.groupby('continent')['country'].nunique()
```

```python
train.groupby('continent')['country'].value_counts()
```

```python
#2. 연도를 기준으로 인구의 최소 최대값을 구하시오.
train.groupby('year')['population'].min()
```

```python
train.groupby('year')['population'].max()
```

```python
#min(최소), max(최대), mean(평균), std(표준편차), var(분산)
#quantile(퍼센트)
```

```python
train['year'].quantile(0.25) #=> 이상치판단 시에 이 함수 필요!
```

```python
train.groupby('year')['population'].agg([min, max])
#여기서의 agg는 무슨 함수일까요? -> 여러개의 함수를 한꺼번에 실행할 때 쓰는 것(Aggregate)
```
